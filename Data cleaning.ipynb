Merge all data into one file

```python
all_data = pd.concat([aldi,tj,fwm,ft,wfm], axis=0)
```

```python
all_data.head()
```

```python
all_data.count()
```

```python
all_data.to_csv('Merged_data_20221123.csv')
```

```python
all_data = all_data.dropna(subset=['Price'])
```

```python
all_data['Price'] = [x.split('for ')[-1] for x in all_data['Price']]
#finding non numerical price row
all_data['Price'] = pd.to_numeric(all_data['Price'],errors='coerce')
all_data['Date'] =  pd.to_datetime(all_data['Date'], infer_datetime_format=True)
all_data['Category'] = all_data['Category'].str.replace('dairy-eggs','dairy')
all_data['Category'] = all_data['Category'].astype('str') 
```

```python
all_data['Kosher'] = [0] * 8572
all_data['Pescatarian'] = [1] * 8572
all_data['No red meat'] = [1] * 8572
all_data['Dairy-free'] = [1] * 8572
all_data['Gluten-free'] = [1] * 8572
```

```python
# Apply Kosher rules
for i in range(0, 8572):
  if all_data['Category'].iloc[i] == 'dairy':
    all_data['Kosher'].iloc[i] = 1
  else:
    if all_data['Category'].iloc[i] == 'produce':
      all_data['Kosher'].iloc[i] = 1
    else:
      if all_data['Category'].iloc[i] == 'seafood':
        if 'fish' in all_data['Item name'].iloc[i].lower():
          all_data['Kosher'].iloc[i] = 1
      else:
        if 'lamb' in all_data['Item name'].iloc[i].lower():
          all_data['Kosher'].iloc[i] = 1
        elif 'beef' in all_data['Item name'].iloc[i].lower():
          all_data['Kosher'].iloc[i] = 1
```

```python
# apply pescatarian rules
for i in range(0, 8572):
  if all_data['Category'].iloc[i] == 'meat':
    all_data['Pescatarian'].iloc[i] = 0
```

```python
# apply no red meat rules
for i in range(0, 8572):
  if all_data['Category'].iloc[i] == 'meat':
    if 'lamb' in all_data['Item name'].iloc[i].lower():
      all_data['No red meat'].iloc[i] = 0
    else:
      if 'beef' in all_data['Item name'].iloc[i].lower():
        all_data['No red meat'].iloc[i] = 0
      else:
        if 'pork' in all_data['Item name'].iloc[i].lower():
          all_data['No red meat'].iloc[i] = 0
        else:
          if 'bacon' in all_data['Item name'].iloc[i].lower():
            all_data['No red meat'].iloc[i] = 0
          else:
            if 'ham' in all_data['Item name'].iloc[i].lower():
              all_data['No red meat'].iloc[i] = 0
            else:
              if 'pepperoni' in all_data['Item name'].iloc[i].lower():
                all_data['No red meat'].iloc[i] = 0
```

```python
# apply dairy free rule
for i in range(0, 8572):
  if all_data['Category'].iloc[i] == 'dairy':
    if 'dairy free' in all_data['Item name'].iloc[i].lower():
          all_data['Dairy-free'].iloc[i] = 1
    else:
      if 'dairy-free' in all_data['Item name'].iloc[i].lower():
        all_data['Dairy-free'].iloc[i] = 1
      else:
          if 'oat' in all_data['Item name'].iloc[i].lower():
            all_data['Dairy-free'].iloc[i] = 1
          else:
            if 'soy' in all_data['Item name'].iloc[i].lower():
              all_data['Dairy-free'].iloc[i] = 1
            else:
              if 'almond' in all_data['Item name'].iloc[i].lower():
                all_data['Dairy-free'].iloc[i] = 1
              else:
                all_data['Dairy-free'].iloc[i] = 0
```

```python
# apply gluten free rule
for i in range(0, 8572):
  if all_data['Category'].iloc[i] == 'dairy':
    if 'granola' in all_data['Item name'].iloc[i].lower():
          all_data['Gluten-free'].iloc[i] = 0
```

```python
all_data.to_csv('Merged_clean_data_20221123.csv')
```
